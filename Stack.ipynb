```python
import pandas as pd
```

```python
df = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
df
```

Stacking:

```python
# transfers columns to rows
df.stack()
# technically a 1-d object
```

Unstacking:

```python
st = df.stack()
```

```python
st.unstack()
```

```python
st.unstack().unstack()
```

```python
st.unstack(0)
#st.unstack("country")
```

```python
st.unstack(["country", "year"])
#st.unstack([2,0])
```

Pivoting:

```python
sales = pd.read_csv("sales.csv", parse_dates = ["Date"])
sales["Salesman"] =  sales["Salesman"].astype("category")
```

```python
sales["Salesman"].value_counts()
# structure is optimal for pivoting
# wnat to create column headers for each salesman
```

```python
sales.pivot(index="Date", columns = "Salesman", values="Revenue")
```

Create an aggregate summary:

```python
food = pd.read_csv("foods.csv")
food
```

```python
food.pivot_table(values= "Spend", index = "Gender", aggfunc = "mean")
```

```python
food.pivot_table(values= "Spend", index = "City", aggfunc = "mean")
```

```python
food.pivot_table(values= "Spend", index = ["Gender", "Item"], aggfunc = "mean")
```

```python
food.pivot_table(values= "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "mean")
```

```python
food.pivot_table(values= "Spend", index = ["Gender", "Item"], columns = ["City", "Frequency"], aggfunc = "mean")
```

```python
food.pivot_table(values= "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "count")
#pd.pivot_table(data = food, values= "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "count")
```

Melting:

```python
quart = pd.read_csv("quarterlysales.csv")
quart
```

```python
nq = quart.melt(id_vars = "Salesman", var_name = "Quarter",  value_name = "Revenue")
nq
```

```python
nq = nq.set_index(["Salesman", "Quarter"])
nq.sort_index()
```
