```python
import pandas as pd
```

```python
# datatype efficiency:
employees = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
# note: today's date gets attached to times without date
employees["Gender"] = employees["Gender"].fillna('NB').astype("category")
```

```python
employees.info()
```

```python
employees[employees["Gender"] == "Female"]
```

```python
# will turn NaN into true
employees["Senior Management"] = employees["Senior Management"].astype("bool")
employees[employees["Senior Management"]]
```

```python
employees[employees["Salary"] > 140000]
```

```python
mask1 = employees["Salary"] > 140000 
mask2 = employees["Gender"] == "Female"
employees[mask1 & mask2]
# employees[mask1 | mask2]
# parenthesis for avoiding ambiguity: (mask1 & mask2) | mask3, etc.
```

```python
employees["Team"].isin(["Sales", "Finance", "Legal"])
employees["Team"].isnull()
employees["Team"].notnull()
employees["Salary"].between(70000, 100000) #inclusive
employees["Last Login Time"].between("8:00AM", "12:00PM") #inclusive
```

```python
# duplicate rules do apply to NaN values
employees.duplicated()
~employees.duplicated(keep=False)
```

```python
employees.drop_duplicates(subset=["First Name"], keep = 'first')
```

```python
# list unique values or count the number of them
employees["Team"].unique()
employees["Team"].nunique(dropna= False) 
```
