Iterating through groups:

```python
starter = pd.DataFrame(columns = df.columns)
starter
```

```python
# append the n largest revenue to the starter DataFrame
for sector, data in sectors:
    high_rev_in_gp = data.nlargest(2,"Revenue")
    starter = starter.append(high_rev_in_gp)
```

```python
starter
```

```python
cities = df.groupby("Location")
starter = pd.DataFrame(columns = df.columns)
```

```python
for city, data in cities:
    most_emp_per_gp = data.nlargest(2,"Employees")
    starter = starter.append(most_emp_per_gp)    
```
